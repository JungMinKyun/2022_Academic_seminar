```python
from xverse.transformer import MonotonicBinning
from xverse.transformer import WOE
```

```python
df_card = pd.read_csv('/content/drive/MyDrive/학술제/카드사기 분류/creditcard.csv')
```

```python
features = [ 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28' ]
```

```python
df_card_X = df_card[ features ]
```

```python
df_card_y = df_card.Class
```

BInning

```python
clf = MonotonicBinning()
clf.fit( df_card_X, df_card_y )

print( clf.bins )
```

woe 계산

```python
clf = WOE()
clf.fit( df_card_X, df_card_y )

print( clf.woe_df )
```

iv값 계산

```python
print(clf.iv_df)
```

```python
iv_value = pd.DataFrame( clf.iv_df )
```

```python
iv_value.reset_index( drop = True, inplace = True )
```

```python
iv_value
```

iv가 0.5 이상

```python
feature_selection_result = list( iv_value['Variable_Name'][ iv_value['Information_Value'] >= 0.5 ] )
```
